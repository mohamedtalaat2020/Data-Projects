```python
#_____FIRST METHOD________________________Removing samples contains missing data_____________________________________________
```

```python
import pandas as pd
data=pd.read_csv('2000.csv',na_values=['undefined'])
mask=data.isnull().any(axis=1)
#calculate how many row contain null value 
num_rows_with_nan=mask.sum()
#calculate how many row in the dataset
total_num_of_rows=len(data)
#calculate the average of null values in the dataset
num_rows_with_nan/total_num_of_rows
#result is 0.22222 so it's small and we can remove these samples
```

```python
data_cleaned=data[~mask]
data_cleaned
```

```python
import pandas as pd
data=pd.read_csv('3000.csv',na_values=['?','undefined'])
mask=data.isnull().any(axis=1) #only rows
num_rows_with_nan=mask.sum()
total_num_of_rows=len(data)
num_rows_with_nan/total_num_of_rows
#result is 0.56 so it's very big and we cannot remove these samples because we will lose alot of rows in the dataset
#so 1 - we will see first number of column  that contains null and how many null value in each column
#   2 - then if there are one column contain alot of null values we will remove it
#   3 - then we will remove the samples that contain null values
```

```python
# 1 - we will see first number of column  that contains null and how many null value in each column
num_cols_with_nan=data.isnull().any(axis=0).sum()
num_cols_with_nan
#show how many column have null value
```

```python
data.isnull().sum()
#show how many nan value in each column if there are column contain alot of null values we can remove it
```

```python
#show avrage of null values in each column
data.isnull().sum()/len(data)
```

```python
# 2 - then if there are one column contain alot of null values we will remove it
#drop column that contain alot of null values
new_data=data.drop(columns=['x3'])

#   3 - then we will remove the samples that contain null values
#calculate number of rows contains null values 
mask=new_data.isnull().any(axis=1)
num_of_rows=mask.sum()

#calculate average(نسبة) of rows contains null values
num_of_rows/len(new_data)
data_clean=new_data[~mask]
#show data after cleaning
data_clean
```

```python
#_____SECOND METHOD________________________filling missing values (IMPUTATION)_____________________________________________
```

```python
# 1- filling null values with  zero
import pandas as pd
data=pd.read_csv('1000.csv',na_values=['undefined','?'])
data_cleaned=data.fillna(0.0)
data_cleaned
data
```

```python
# 2- filling null values with  choosen values
import pandas as pd
data=pd.read_csv('1000.csv',na_values=['undefined','?'])
data_cleaned=data.fillna({'x1':0,'x3':1,'x6':2 })
data_cleaned
```

```python
# 3- filling null values with  mean,median,mode

import pandas as pd
data=pd.read_csv('1000.csv',na_values=['undefined','?'])
data_cleaned=data.fillna(data.mean(axis=0))
data_cleaned
```

```python
# 3- filling null values with median

import pandas as pd
data=pd.read_csv('1000.csv',na_values=['undefined','?'])
data_cleaned=data.fillna(data.median(axis=0))
data_cleaned
```

```python
# 3- filling null values with mode (using with categorical data)

import pandas as pd
data=pd.read_csv('4000.csv',na_values=['undefined','?'])
#data_cleaned=data.fillna(data.mode())
#data_cleaned
#data.mode()
data.mode(axis=0)
#cleaning categorical data
most_frequent=data['x3'].mode()[0]
cleaned_data1=data.fillna({'x3':most_frequent})
cleaned_data1
#after cleaning categorical data we can clean all data
data_cleaned2=cleaned_data1.fillna(cleaned_data1.mean(axis=0))
data_cleaned2
```
