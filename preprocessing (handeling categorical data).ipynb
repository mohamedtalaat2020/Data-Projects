```python
import pandas as pd
data=pd.read_csv('5000.csv')
data
#use unique funcation to get categorical data in each column

```

```python
print(data['x1'].unique())
print(data['x3'].unique())
print(data['x6'].unique())
```

```python
#numeric encoding for categorical data
encoded_data=data.replace({
   'x1' : {'male':0 ,'female' :1 },
   'x3' : {'low':0 ,'med' :1 ,'high' :2 },
   'x6' : {'full-time':0 ,'unemployed': 1,'part-time' :2 }
    })
encoded_data
```

```python

#one-hot encoding for categorical data(better for nominal data)
encoded_data1=data.replace({
   'x1' : {'male':0 ,'female' :1 },
   'x3' : {'low':0 ,'med' :1 ,'high' :2 },
    })
encoded_data2=pd.get_dummies(encoded_data1)
encoded_data2


```
